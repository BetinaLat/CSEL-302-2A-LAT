```python
import pandas as pd

data = {
    'student_id': range(1,21),
    'gender': ['Male','Female','Female','Male','Male','Male','Female','Male','Female','Male',
               'Female','Male','Male', 'Female','Female','Female','Male','Male','Male','Male'],
    'test_date': pd.date_range(start='2023-09-25',periods=20),
    'study_hours': [2,3,5,2,6,4,2,5,3,5,3,2,3,5,6,3,5,3,5,3],
    'test_score': [75,82,90,88,65,73,79,80,93,98,65,82,85,87,90,75,84,86,90,94],
    'attendance': ['Present', 'Present','Absent','Present','Present','Present','Present','Absent','Present','Present',
                   'Present','Present','Absent','Present','Present','Present','Present','Absent','Present','Present',]
}
df = pd.DataFrame(data)
df.to_csv('student_performance.csv', index=False)
```

```python
df = pd.read_csv('student_performance.csv')
print(df.head(8))
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sne.histplot(df['study_hours'],bins=15,kde=True)
plt.title('Distribution of Test Score')
plt.subplot(1,2,2)
sne.histplot(df['study_hours'],bins=15,kde=True)
plt.title('Distribution of Study Hours')
plt.figure(figsize=(8,6))
sne.scatterplot(data=df,x='study_hours', y='test_score')
plt.title('Scatter Plot of the Study Hours vs. Test Scores')
```

```python
plt.figure(figsize=(6,4))
sne.countplot(data=df,x='gender')
plt.title('Gender Distribution')
plt.figure(figsize=(30,18),facecolor='#78bfff')
sne.lineplot(data=df,x='test_date', y='test_score',marker='o')
plt.title('Trend in Test Score Over Time')
plt.grid(axis='both')
correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()
plt.figure(figsize=(8,6))
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```
